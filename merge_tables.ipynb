```python
import pandas as pd
```

```python
df = pd.read_csv('coordinates.csv', index_col=0)
cities = pd.read_csv('cities.csv', index_col=0)
```

```python
df.dtypes
```

```python
longitude_list = []
latitude_list = []
for i, longitude in df['Longitude'].items():
  longitude_list.append(longitude.replace(',', '.'))

for i, latitude in df['Latitude'].items():
  latitude_list.append(latitude.replace(',', '.'))
df['Longitude'] = longitude_list
df['Latitude'] = latitude_list
```

```python
df['Longitude'] = pd.to_numeric(df['Longitude'])
df['Latitude'] = pd.to_numeric(df['Latitude'])
df
```

```python
city_list = []
for i, city in df['city'].items():
  city_list.append(city.replace(' ', '-').replace('ą', 'a').replace('ć', 'c').replace('ę', 'e').replace('ł', 'l') \
               .replace('ń', 'n').replace('ó', 'o').replace('ś', 's').replace('ż', 'z').replace('ź', 'z').replace('""',
                                                                                                                  ''))
df['city'] = city_list
```

```python
result = pd.merge(cities, df, on=['city', 'city'], validate='m:m')
```

```python
result.to_csv('source.csv', index=True)
```
